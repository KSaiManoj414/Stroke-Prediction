```python
#Imputing Missing/NA Values in dataset using Attribute Mean/Median
data['bmi'].fillna(np.round(data['bmi'].median(), 1), inplace = True) ##replaced NA values in BMI attr with Median

#Printing top 5 rows to check consistency of data
data.head(5)
print(f"After imputing Null values of the Dataset :")
data.isna().sum().to_frame()
```

```python
##Removed rows  where "Age= Other"
data['age'] = data['age'].astype('int')
data = data[data['gender'] != 'Other']
```

```python
#Preparing X & Y data, and the train - test split
## A test set of 30% of dataset is assumed for this model development
x=data.drop(['stroke'], axis=1)
y=data['stroke']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 124)
```

```python
numerical_variables = ['age', 'hypertension', 'heart_disease','avg_glucose_level','bmi']
categorical_variables = ['gender','ever_married','work_type','Residence_type','smoking_status']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numerical_pipeline = Pipeline([('impute_missing', SimpleImputer( strategy = 'median')),
                              ('standardize_num', StandardScaler())])
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_pipeline = Pipeline([('impute_missing_cats', SimpleImputer( strategy = 'most_frequent')),
                               ('create_dummies_cats', OneHotEncoder(handle_unknown = 'ignore', drop = 'first'))])
```

```python
#https://github.com/appliedecon/data602-lectures/blob/main/sklearn-pipeline-example/credit.ipynb
from sklearn.compose import ColumnTransformer

categorical_pipeline_processing = ColumnTransformer( transformers = [('create_dummies', categorical_pipeline, categorical_variables)])
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler.fit_transform
categorical_pipeline_processing.fit_transform(x_train, y_train)
```

```python
# insert code here
#https://github.com/appliedecon/data602-lectures/blob/main/sklearn-pipeline-example/credit.ipynb
processing_pipeline = ColumnTransformer(transformers=[('proc_numeric', numerical_pipeline, numerical_variables), 
                                                      ('create_dummies', categorical_pipeline, categorical_variables)])
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler.fit_transform
processing_pipeline.fit_transform(x_train, y_train)
```

```python
print(f'Training examples: {x_train.shape[0]}')
print(f'Test examples: {x_test.shape[0]}')
```

```python
#Decision Tree Classifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
dct_clf = DecisionTreeClassifier(criterion = 'entropy',max_depth=2)

pipeline_modeling = Pipeline([('data_processing', processing_pipeline), ('tree', dct_clf)])

param_grid = [
{'tree__max_depth': [2, 5, 10, 15, 20],
'tree__min_samples_split':[3, 5, 10, 20, 40],
'tree__min_samples_leaf': [2, 5],
'tree__class_weight':[None, 'balanced']
}]

gcv_results = GridSearchCV(estimator=pipeline_modeling, param_grid=param_grid, scoring='recall', refit=True)
gcv_results = gcv_results.fit(x_train, y_train)
gcv_results.best_estimator_


#pipeline_modeling.fit(x_train,y_train)


y_pred = gcv_results.predict(x_test)
print("Test Accuracy: {} \n".format(accuracy_score(y_test, y_pred)))

print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))

print("Classification Report: \n", classification_report(y_test, y_pred))
```
